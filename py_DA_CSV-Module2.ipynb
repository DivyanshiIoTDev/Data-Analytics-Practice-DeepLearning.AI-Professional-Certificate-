```python
#Filtering data based on one dataset needed
 
#Finding unique reponses of Rating
unique_rating=df["Rating"].unique()
print(unique_rating)
```

```python
for unique in unique_rating:  #Filtering includes dataframe including rows which has rating as >=4.5
    if unique>=4.5:
        high_rating=(df[df["Rating"]==unique])
        print(high_rating)
```

```python
#SETTING INDEXES AS A Particular series values of a column
INDEX_AS_ZIPCODE=df.set_index("Zip Code")
print(INDEX_AS_ZIPCODE)
```
